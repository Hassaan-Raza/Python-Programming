```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
our_data = pd.read_csv('diabetes.csv')
```

```python
X = our_data.drop(columns= 'Outcome', axis=1)
Y = our_data['Outcome']
```

```python
print(X)
print(Y)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
```

```python
Y=our_data['Outcome']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
# training the support vector Machine Classifier
classifier.fit(X_train,Y_train)
```

```python
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```

```python
input_data=(5,44,62,0,0,25,0.587,36)

#now we change this input into a numpy array

input_data_array = np.asarray(input_data)
```

```python
#now reshaping array as we are only predicting for one instance

input_data_reshape = input_data_array.reshape(1,-1)

# this will tell the model that we are not giving 768 examples but we are just trying to predict the label for only one instance

#standaraized data
input_data_standardized = scaler.transform(input_data_reshape) 
```

```python
prediction = classifier.predict(input_data_standardized)
if(prediction[0]==0):
    print('The Person is not diabetic')
else:
    print('The Person is diabetic')
```
